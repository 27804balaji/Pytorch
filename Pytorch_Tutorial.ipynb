## Slicing

* `list[start : stop : step]` in Python

```python
lst = [10, 20, 30, 40, 50]

print(lst[1:4])   # [20, 30, 40] (Indexes 1 to 3)
print(lst[:3])    # [10, 20, 30] (From start to index 2)
print(lst[2:])    # [30, 40, 50] (From index 2 to end)
print(lst[:])     # [10, 20, 30, 40, 50] (Full list)
```

## Slice with steps

```python
lst = [10, 20, 30, 40, 50]

print(lst[::2])    # [10, 30, 50] (Every 2nd element)
print(lst[1::2])   # [20, 40] (Start from index 1, every 2nd element)
print(lst[::-3])   # if the step is negative the  all values are assigned with reverse index value hence 50 has index value 0. from 0th index it select the value which is present in the -3th index. accorrding to our list -3th index is occupied by 20
print(lst[3:0:-1]) # [40, 30, 20] (Reverse from index 3 to 1)
```

# **Tensor**

* **Automatic Differentiation –** Tensors can track gradients(Updating weights) for deep learning models using requires_grad=True.
* **Efficient Operations –** PyTorch provides optimized tensor operations for machine learning and deep learning applications.

## Matrix multiplication

Two main type of performing multiplication ion matrix

1. Element wise multiplication
2. matrix multiplication


There are Two main rules to perform matrix multiplication to be satisfied

1. **Inner dimensions** must match

* `(3, 2) @ (3, 2)` won't work.
* `(2, 3) @ (3, 2)` will work.

2. The resulting matrix has the output shape of **outer dimensions**:

* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`


Note:- Matrix Multiplication -> http://matrixmultiplication.xyz/

```python
torch.rand(3, 2) @ torch.rand(3, 2) # @ -> matmul()
```

```python
torch.rand(3, 2) @ torch.rand(2, 3)
```

```python
# Element wise multiplication.

print(tensor, "*", tensor)
print(f"Equals : {tensor * tensor}")
```

```python
# Matrix multiplication.

print(tensor, "*", tensor)
print(f"Equals : {torch.matmul(tensor, tensor)}")
```

# One of the common error in deep learning : Shape error.

```python
# Shape of the tensor.

import torch

tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

# print(tensor_A.shape, tensor_B.shape)
try:
  torch.mm(tensor_A, tensor_B) # torch.mm() -> torch.matmul()

except RuntimeError:
  print(f"The Shape {tensor_A.shape} and {tensor_B.shape} cann't be multiplied.")
```

* This issue can fixed by one of the method **transponse**.

```python
tensor_B.T # .T -> refers to transpose.
```

```python
torch.mm(tensor_A, tensor_B.T)
```

## Reshaping, Stacking, Squeezing and Unsqueezing tensors

* Reshaping -> reshape a input tensor into defined shape(keeps the same size of the original tensor in the memory).
* Stacking -> combine the multiple tensor on the top of the each other(vstack :- combines the two or more tensors vertically along with the new dimension) or side by side(hstack :- combines the two or more tensors horizontally along with the new dimension)
* Squeezing -> remove all `1` dimension from the tensor(remove all dimension size of 1 from the targeted tensor).
* Unsqueezing -> add a `1` dimension to the target tensor(add dimension size of 1 to the targeted tensor).
* Permute -> rearrange the dimensions of a tensor in a specified order without changing its data.

```python
# Reshaping the tensor.

x = torch.arange(1, 9)
x, x.shape
```

```python
x_reshapped = x.reshape(1, 8)
x_reshapped, x_reshapped.shape
```

```python
# HStacking.

x_stack = torch.stack([x, x, x], dim = 0) # dim = 0 refers to stack along a new row.
x_stack
```

```python
# VStacking

x_stacked = torch.stack([x, x, x], dim = 1) # dim = 1 refers to stack along a new column.
x_stacked
```

```python
# Squeezing.

x = torch.rand(1, 3, 3)
x, x.shape
```

```python
x_squeezed = x.squeeze()
x_squeezed, x_squeezed.shape
```

```python
# Unsqueezing.

x_unsqueezed = x_squeezed.unsqueeze(dim = 0) # dim refers to where to add the dimension size of 0 means it will add the dimension at the 0th index.
x_unsqueezed, x_unsqueezed.shape
```

```python
# Permuting.

x_image = torch.rand(254, 254, 3)
x_image.shape
```

```python
x_permuted = x_image.permute(2, 0, 1) # need to mention the index that we need to rearange the dimension.
x_permuted.shape
```
